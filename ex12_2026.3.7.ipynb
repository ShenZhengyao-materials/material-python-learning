```python
"""3.创建两个 3×2 和 2×3 矩阵，做合法矩阵乘法。"""
import numpy as np
arr1 = np.random.randint(1,9,size=(3,2))
print(arr1)
arr2 = np.random.randint(1,9,size=(2,3))
print(arr2)
a = np.dot(arr1,arr2)
print(a)
```

```python
"""4.生成 5×5 矩阵，把大于平均值的元素全部设为 0。"""
import numpy as np
arr = np.random.random(25)
b = np.mean(arr)
arr = arr.reshape(5,5)
print(arr)
c = arr > b
print(c)
arr[c] = 0
print(arr)
```

```python
"""2.模拟数据处理：
生成一组模拟的材料性能数据（100 个样本，5 个特征，用 np.random.rand(100,5)），计算：
每个特征的最大值、最小值、标准差；
筛选出 “第 1 个特征> 0.5 且第 2 个特征 < 0.3” 的样本。"""
#import numpy as np
#arr = np.random.rand(10,5)
#print(arr)
#n = 0
#for n in range(10):
#c = arr[n][0] > 0.5
#d = arr[n][1] < 0.3
#print(arr[n][c][d])

#修改:
import numpy as np
# 1. 生成100个样本、5个特征的模拟材料性能数据
material_data = np.random.rand(100, 5)
print("模拟材料数据（前5行）：\n", material_data[:5])

# 2. 计算每个特征的最大值、最小值、标准差（axis=0 按列计算）
max_vals = material_data.max(axis=0)
min_vals = material_data.min(axis=0)
std_vals = material_data.std(axis=0)

print("\n每个特征的最大值：", max_vals)
print("每个特征的最小值：", min_vals)
print("每个特征的标准差：", std_vals)

# 3. 筛选：第1个特征>0.5 且 第2个特征<0.3（NumPy布尔掩码，不用循环！）
# 注意：Python/NumPy索引从0开始，“第1个特征”是[:,0]，“第2个特征”是[:,1]
mask = (material_data[:, 0] > 0.5) & (material_data[:, 1] < 0.3)
filtered_data = material_data[mask]

print("\n符合条件的样本数：", len(filtered_data))
print("符合条件的样本：\n", filtered_data)
```

```python
"""3.晶格常数计算：
给定一个 3×3 的晶格矩阵（比如简单立方：[[a,0,0],[0,a,0],[0,0,a]]），用 np.linalg.det 计算晶胞体积，用 np.linalg.eig 求晶格矩阵的特征值（验证晶格常数）。"""
import numpy as np
a = 3
arr = np.array([[a,0,0],[0,a,0],[0,0,a]])
print(arr)
v = np.linalg.det(arr)
print(v)
r = np.linalg.eig(arr)
print(r)
```
