```python
import pandas as pd
import numpy as np
data = {
    'name': ['张三','李四','王五','赵六'],
    'score': [85, 92, 78, 90],
    'class': ['A','B','A','B']
}
df = pd.DataFrame(data)
```

```python
"""1.选出 class=A 的人
2.算出所有 score 的平均分
3.按 score 从高到低排序
4.给每个人加一列：是否及格（score>=60）"""
df1 = df[df['class'] == 'A']
print("一.",df1)
df2 = df["score"]
arr = np.array(df2)
#print(arr)
av = np.mean(arr)
print("二.",av)
df3 = df.sort_values("score",ascending=False)
print("三.",df3)
df["是否及格"] = df["score"] >= 60
print(df)
```

```python
import pandas as pd
import numpy as np

# 1. 初始化数据（固定原始数据，避免后续操作污染）
data = {
    'name': ['张三','李四','王五','赵六'],
    'score': [85, 92, 78, 90],
    'class': ['A','B','A','B']
}
df = pd.DataFrame(data)

# 2. 任务1：选class=A的人（仅展示，不修改原始df，无是否及格列）
df1 = df[df['class'] == 'A']
print("一. class=A的人（无是否及格列）：")
print(df1)
print("-" * 40)  # 分隔线，方便看结果

# 3. 任务2：算平均分（仅输出数值，不修改df）
df2 = df["score"]
arr = np.array(df2)
av = np.mean(arr)
print("二. 所有score的平均分：", av)
print("-" * 40)

# 4. 任务3：按score降序排序（仅展示排序结果，无是否及格列）
df3 = df.sort_values("score", ascending=False)
print("三. 按score降序排序（无是否及格列）：")
print(df3)
print("-" * 40)

# 5. 任务4：给原始df加「是否及格」列，保持原始行顺序，仅这一步展示该列
df["是否及格"] = df["score"] >= 60  # 新增列，行顺序不变
print("四. 原始顺序+是否及格列：")
print(df)
```
