# 一、ndarray
## 1.ndarray的特性
### 多维性

```python
import numpy as np
```

```python
arr = np.array(1)
print(arr)
print("arr的维度：", arr.ndim)
```

```python
arr = np.array([1, 2, 3])
print(arr)
print("arr的维度：", arr.ndim)
```

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr)
print("arr的维度：", arr.ndim)
```

### 同质性

```python
arr = np.array([1, "hello"]) #强制转化
print(arr)
```

```python
arr = np.array([1, 2.5])
print(arr)
```

## 2.ndarray的属性

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr)
print("数组的形状",arr.shape) #2行3列
```

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr)
print("数组的维度",arr.ndim)
```

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr)
print("元素的个数",arr.size)
```

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr)
print("数据类型",arr.dtype)
```

```python
arr = np.array([[1, 2.5, 3],[4, 5, 6]])
print(arr)
print("数据类型",arr.dtype)
```

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr)
print("元素的转置",arr.T)
```

## 3.ndarray的创建

```python
list1 = [1, 2, 3] #基础
arr = np.array(list1)
print(arr)
```

```python
arr1 = np.copy(arr)
print(arr1)
arr1[0] = 9
print(arr1)
print(arr)
```

```python
#预定义形状
#全0 全1 未初始化 固定值 类似
arr = np.zeros((3,3),dtype = np.int64)
print(arr)
```

```python
arr = np.ones((3,3),dtype = np.int64)
print(arr)
```

```python
arr = np.empty((3,3))
print(arr)
```

```python
arr = np.empty((10,2))
print(arr)
```

```python
arr = np.full((4,2),2026)
print(arr)
```

```python
arr1 = np.zeros_like(arr)
print(arr1)
```

```python
arr1 = np.empty_like(arr)
print(arr1)
```

```python
# 等差数列
arr = np.arange(1,10,2) # start,end(except),step
print(arr)
```

```python
# 等间隔数列
arr = np.linspace(1,10,5) # start,end,份数(平均分配)
print(arr)
```

```python
# 对数间隔数列
arr = np.logspace(0,4,2,base=2)
print(arr)
```

```python
arr = np.logspace(0,4,3,base=2)
print(arr)
```

```python
arr = np.logspace(0,4,3,base=3)
print(arr)
```

```python
# 矩阵的构造
arr = np.eye(3) #三阶
print(arr)
```

```python
arr = np.diag([1,2,3,4,5])
print(arr)
```

```python
# 随机数组的生成
arr = np.random.rand(2,3) # 0-1
print(arr)
```

```python
arr = np.random.uniform(3,6,(2,3)) # 范围，(形状)
print(arr)
```

```python
arr = np.random.randint(3,30,(2,3))
print(arr)
```

```python
# 正态分布
arr = np.random.randn(2,3)
print(arr)
```

```python
# 设置随机种子
np.random.seed(20)
arr = np.random.randint(1,10,(2,5))
print(arr)
```
