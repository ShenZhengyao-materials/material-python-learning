```python
""""3.数组运算：创建两个形状为 (2,2) 的随机整数数组（取值范围 1-10），分别实现它们的逐元素加法、减法、乘法、矩阵乘法，并输出结果。"""
import numpy as np
arr1 = np.random.randint(1,11,size=(2,2))
arr2 = np.random.randint(1,11,size=(2,2))
print(arr1)
print(arr2)
print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(np.dot(arr1,arr2))

```

```python
"""4.数组变形与拼接：
将第 1 题的 (3,4) 数组变形为 (4,3) 的数组；
创建一个形状为 (3,4)、元素全为 1 的数组，将其与第 1 题的数组按行拼接（最终形状为 (6,4)），再按列拼接（最终形状为 (3,8)）。"""
import  numpy as np
arr = np.arange(0, 12).reshape((4,3))
print(arr)
arr1 = np.ones((3,4))
print(arr1)
arr3 = np.arange(0, 12).reshape((3,4))
arr4 = np.vstack([arr1,arr3])
arr5 = np.hstack([arr1,arr3])
print(arr4)
print(arr5)
```

```python
"""5.统计与条件筛选：基于第 1 题的数组，完成以下操作：
计算数组的最大值、最小值、平均值、总和；
筛选出数组中大于 5 的所有元素，并输出这些元素的位置（索引）；
将数组中小于 3 的元素替换为 0，大于等于 3 的元素保持不变。"""
import  numpy as np
arr = np.arange(0, 12).reshape((3,4))
print(arr)
a = np.max(arr)
print(a)
b = np.min(arr)
print(b)
c = np.average(arr)
print(c)
d = np.sum(arr)
print(d)
e = arr > 5
print(e)
print(arr[e])
f = np.where(arr >= 5)
print(f)
g = arr < 3
print(g)
arr[g] = 0
print(arr)
```

```python
"""2.用 np.arange 生成 1～10，求平均值与总和。"""
import numpy as np
arr = np.arange(11, dtype=np.int32)
a = np.average(arr)
print(a)
b = sum(arr)
print(b)
```
